```python
# Graph Representation: Adjacency List
graph ={
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def dfs(graph, start, goal):
    try:
        # Check if the start or goal is not in the graph
        if start not in graph or goal not in graph:
            raise ValueError(f"Start node '{start}' or goal node '{goal}' is not in the graph.")

        visited = set()  # To track visited nodes
        stack = [start]  # Stack for DFS
        result = []  # To store the path from start to goal

        while stack:
            node = stack.pop()

            if node == goal:
                result.append(node)
                return result  # Return the path as soon as we find the goal

            if node not in visited:
                visited.add(node)
                result.append(node)  # Add the current node to the path
                stack.extend(reversed(graph[node]))  # Add adjacent nodes to stack

        return f"No path found from {start} to {goal}."  # If goal is not found

    except ValueError as ve:
        return str(ve)  # Return the error message if nodes are not in the graph

    except Exception as e:
        return f"An error occurred: {str(e)}"  # Catch any other unexpected errors

def main():
    start_node = 'A'
    goal_node = 'F'

    print(f"Running DFS from {start_node} to {goal_node}...")
    result = dfs(graph, start_node, goal_node)
    
    if isinstance(result, list):
        print(f"DFS: Path from {start_node} to {goal_node} found!")
        print(f"Path: {' -> '.join(result)}")
    else:
        print(result)  # Print the error message if no path is found or error occurs

# Entry point of the script
if __name__ == "__main__":
    main()

```
